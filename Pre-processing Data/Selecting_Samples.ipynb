```python
import pandas as pd
df = pd.read_csv("clean_pathological_data.csv")
```

```python
df.columns = (
    df.columns
      .str.strip()
      .str.lower()
      .str.replace(" ", "_")
      .str.replace("(", "")
      .str.replace(")", "")
)
```

```python
df.columns
```

```python
df["strata"] = (
    df["primary_tumor_site"].astype(str) + "_" +
    df["grading_hpv"].astype(str) + "_" +
    df["pt_stage"].astype(str) + "_" +
    df["pn_stage"].astype(str)
)
display(df)
```

```python
sampled = (
    df.groupby("strata", group_keys=False)
      .apply(lambda x: x.sample(n=min(len(x), max(1, round(100 * len(x) / len(df)))), random_state=42))
)
```

```python
print("Primary Tumor Site:")
print("Original distribution:")
print(df["primary_tumor_site"].value_counts(normalize=True))

print("\nSampled distribution:")
print(sampled["primary_tumor_site"].value_counts(normalize=True))

print()
print("Grading")

print("Original distribution:")
print(df["grading_hpv"].value_counts(normalize=True))

print("\nSampled distribution:")
print(sampled["grading_hpv"].value_counts(normalize=True))

print()
print("PT Stage")

print("Original distribution:")
print(df["pt_stage"].value_counts(normalize=True))

print("\nSampled distribution:")
print(sampled["pt_stage"].value_counts(normalize=True))

print()
print("PN Stage")

print("Original distribution:")
print(df["pn_stage"].value_counts(normalize=True))

print("\nSampled distribution:")
print(sampled["pn_stage"].value_counts(normalize=True))
```

```python
from scipy.stats import entropy

def kl(col):
    orig = df[col].value_counts(normalize=True)
    samp = sampled[col].value_counts(normalize=True)
    orig, samp = orig.align(samp, fill_value=0)
    return entropy(orig, samp)

for col in ["primary_tumor_site", "grading_hpv", "pt_stage", "pn_stage"]:
    print(col, kl(col))
```
