```python
import requests
import pandas as pd

base = "https://hancock.research.fau.eu/include/get_data.php?tab="

df_path = pd.DataFrame(requests.get(base + "pathological_data").json())
path_ids = set(df_path["patient_id"].astype(str))
```

```python
dup_slide_ids = ['36' '239' '251' '277' '428' '489' '589' '732']
no_slide_ids = ['356', '96', '348', '341', '173', '204', '129', '197', '282', '742', '299', '266', '300', '497', '272', '571', '754', '186', '535', '508', '686', '107', '438', '265', '27', '132', '268', '448', '526', '633', '541', '446', '537', '97', '143', '150', '710', '515', '329', '580', '659', '748', '629', '332', '302', '67', '241', '305', '596', '199', '91', '381', '54', '395', '13', '729', '382', '76', '89', '440', '598', '759']
dup_slide_ids = set(map(str, dup_slide_ids))
no_slide_ids = set(map(str, no_slide_ids))
```

```python
bad_ids = dup_slide_ids | no_slide_ids
clean_path_ids = path_ids - bad_ids
```

```python
print("Original pathological IDs:", len(path_ids))
print("Bad IDs removed:", len(bad_ids & path_ids))
print("Final clean IDs:", len(clean_path_ids))
```

```python
df_clean = df_path[df_path["patient_id"].astype(str).isin(clean_path_ids)]
df_clean.to_csv("clean_pathological_data.csv", index=False)
```
